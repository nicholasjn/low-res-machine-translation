```python
import json
from datasets import load_dataset

ds = load_dataset("indonlp/NusaX-MT", trust_remote_code=True)

TARGET_PAIRS = [
    ("eng", "bbc"),
    ("ind", "bbc"),
]

LANG_TO_NLLB = {
    "eng": "eng_Latn",
    "ind": "ind_Latn",
    "bbc": "bbc_Latn",
}

for src_lang, tgt_lang in TARGET_PAIRS:
    for split in ds:
        output_path = f"sentence_pairs_data/nusax_{src_lang}_to_{tgt_lang}_{split}.jsonl"
        count = 0

        with open(output_path, "w", encoding="utf-8") as f:
            for row in ds[split]:
                item = None

                if row["text_1_lang"] == src_lang and row["text_2_lang"] == tgt_lang:
                    item = {
                        "src_lang": LANG_TO_NLLB[src_lang],
                        "tgt_lang": LANG_TO_NLLB[tgt_lang],
                        "source": row["text_1"],
                        "target": row["text_2"],
                        "source_dataset": "NusaX-MT",
                        "split": split,
                        "id": row["id"],
                    }

                elif row["text_1_lang"] == tgt_lang and row["text_2_lang"] == src_lang:
                    item = {
                        "src_lang": LANG_TO_NLLB[src_lang],
                        "tgt_lang": LANG_TO_NLLB[tgt_lang],
                        "source": row["text_2"],
                        "target": row["text_1"],
                        "source_dataset": "NusaX-MT",
                        "split": split,
                        "id": row["id"],
                    }

                if item is not None:
                    f.write(json.dumps(item, ensure_ascii=False) + "\n")
                    count += 1

        print(f"Saved {count:,} rows to {output_path}")
```
